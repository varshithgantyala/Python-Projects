```python
median_age = df['Age'].median()
```

```python
median_age
```

```python
df.loc[df['Age'].isnull(), 'Age'] = median_age
```

```python
df
```

```python
mode_embarked = df['Embarked'].mode()[0]

# Fill missing values
df.loc[df['Embarked'].isnull(), 'Embarked'] = mode_embarked
```

```python
df
```

```python
df.drop('Cabin', axis=1, inplace=True)
```

```python
df
```

```python
df['FamilySize'] = df['SibSp'] + df['Parch'] + 1
#+1 used for including themselves
```

```python
df
```

```python
survival = df.groupby('Sex')['Survived'].mean()
```

```python
survival
```

```python
survival.plot(kind='bar', color=['lightgreen', 'lightskyblue'])
plt.title('Survival Rate by Gender', color = 'red')
plt.ylabel('Survival Rate', color = 'red')
plt.xticks(rotation=0) # Keep the x-axis labels horizontal
plt.show()
```
